```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
## Simple Examples

x = np.arange(0,10)
y = np.arange(11,21)
```

```python
a = np.arange(40, 50)
b = np.arange(50, 60)
```

```python
## Plooting using matplotlib

## plt scatter

plt.scatter(x,y,c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D') 
plt.savefig('Test.png')
plt.show()
```

```python
y=x*x
```

```python
## plt plot
plt.plot(x,y, 'r', linestyle='--', linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
#Creating Subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r*--')

plt.subplot(2,2,2)
plt.plot(x,y,'g--')

plt.subplot(2,2,3)
plt.plot(x,y,'bo')
```

```python
# Compute the x and y coordinates for points on a sine curve
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
plt.title("Sine Wave Form")

plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2, 1, 1)

plt.plot(x, y_sin, 'r--')
plt.title('Sine')

plt.subplot(2, 1, 2)
plt.plot(x, y_cos, 'y')
plt.title('Cosine')

plt.show()
```

```python
## Bar Plot

x = [2,8,10]
y = [11,10,9]

x2 = [3,6,11]
y2 = [6,15,7]
plt.bar(x,y)
plt.bar(x2, y2, color = 'g')
plt.title("Bar Graph")

plt.show()
```

```python
# Histograms

a = np.array([22, 87, 5, 43, 24, 15, 11 , 20 , 9, 45, 31, 27])
plt.hist(a)
plt.title("Histogram")
plt.show() 
```

```python
# Box Plot 

data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data, vert=True, patch_artist=True)
```

```python
# Data to Plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)

#Plot 
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
