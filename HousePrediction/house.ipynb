```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns

pd.pandas.set_option('display.max_columns', None)
```

```python
dataset = pd.read_csv('housing.csv')

print(dataset.shape)
```

```python
dataset.head()
```

```python
## 1 - step make the list of features which has missing values 
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum() > 1]

print(features_with_na)
```

```python
## 2 - step print the feature name and the percentage of missing values
for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(), 4), '% missing values')
```

```python
for feature  in features_with_na:
    data = dataset.copy()


    # let's make a varibale that indicates 0 if the observation was missing or 
    data[feature] = np.where(data[feature].isnull(), 1, 0) 
    ## check whether the feature has null values if yes then 1 else 0

    ## let's calculate the mean salesPrice where the info is missing or pre
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
print("Id of Houses {}".format(len(dataset.Id)))
```

```python
# list of numerical variables
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']  # O means Obejct

print('Number of Numerical Variables: ', len(numerical_features))


dataset[numerical_features].head()
```

From the Dataset we have 4 year variables. We have extract information from the datetime like no of years or no. of days. One Example in this specific scenario can be difference in years between the year the house was built and the year the house was sold. 

```python
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

year_feature
 ## Temporal varibales
```

```python
## Content of these year variables

for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
## Lets analyze the Temporal Datetime Varibales
## We will check whether there is a relation between year the house is sold and 

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Year vs YearSold")
```

```python
for feature in year_feature:
    if feature != 'YrSold':
        data = dataset.copy()

        data[feature]=data['YrSold']-data[feature]

        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
discrete_feature= [feature for feature in numerical_features if len(dataset[feature].unique()) < 25 and feature not in  year_feature + ['Id']]
print("Discrete Variables Count: {}".format(len(discrete_feature)))
```

```python
print(discrete_feature)
dataset[discrete_feature].head()
```

```python
## relationship between Discrete Feature and SalePrice
for feature in discrete_feature:
    data = dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```
